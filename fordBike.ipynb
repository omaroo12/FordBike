```python
df['duration_hour']=df['duration_sec']/3600
```

```python
df
```

```python
df.start_time=pd.to_datetime(df.start_time)
df.end_time=pd.to_datetime(df.end_time)
```

```python
df['start_time'].dt.day_name().hist()
```

```python
df['start_time'].dt.hour.hist()
```

```python
import matplotlib.pyplot as plt
```

```python
male_condition=df['member_gender']=='Male'
male_condition.value_counts().plot(kind='bar', color=['blue', 'orange'])

plt.title('Distribution of Member Gender')
plt.xticks([0, 1], ['Male', 'Female'], rotation=0)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
groupByBirthYear=df.groupby('member_birth_year')['duration_hour']
birthYearAgg=groupByBirthYear.agg(['mean','count','max'])
```

```python
birthYearAgg.plot(y='count')
```
